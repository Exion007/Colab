```
import os
import time
import pickle
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, log_loss
from sklearn.ensemble import BaggingClassifier, GradientBoostingClassifier, RandomForestClassifier
```

```
df = pd.read_csv("pokemon.csv")

df["Type 2"].fillna("None", inplace = True)

df = df.iloc[:,2:23]
df = df.drop(columns=["Abilities", "Mean", "Standard Deviation"])
df.info()
```

```
print(df["Type 1"].unique(), "\n")
print(df["Type 2"].unique(), "\n")
print(df["Experience type"].unique(), "\n")
print(np.sort(df["Experience to level 100"].unique()), "\n")
print(np.sort(df["Catch Rate"].unique()), "\n")
```

```
le = LabelEncoder()
df["Type 1"] = le.fit_transform(df["Type 1"])
df["Type 2"] = le.fit_transform(df["Type 2"])
df["Generation"] = le.fit_transform(df["Generation"]) + 1
df["Final Evolution"] = le.fit_transform(df["Final Evolution"])
df["Experience type"] = le.fit_transform(df["Experience type"])
df['Legendary'] = le.fit_transform(df['Legendary'])
df['Mega Evolution'] = le.fit_transform(df['Mega Evolution'])
df['Alolan Form'] = le.fit_transform(df['Alolan Form'])
df['Galarian Form'] = le.fit_transform(df['Galarian Form'])
df.head(10)
```

```
df.to_csv("pokemon_df.csv", index = False)
```
